```python
sub1 = float(input("Enter marks for English: "))
sub2 = float(input("Enter marks for Math: "))
sub3 = float(input("Enter marks for Science: "))
sub4 = float(input("Enter marks for Islamiat: "))
sub5 = float(input("Enter marks for Urdu: "))

total = sub1 + sub2 + sub3 + sub4 + sub5
percentage = (total / 500) * 100

if percentage >= 80:
    grade = "A+"
elif percentage >= 70:
    grade = "A"
elif percentage >= 60:
    grade = "B"
elif percentage >= 50:
    grade = "C"
else:
    grade = "Fail"
print("-----------------------")
print("Total Marks Obtained:", total)
print("Percentage:", percentage, "%")
print("Grade:", grade)
print("-----------------------")
```

```python
num1 = int(input("Enter First Number: "))
op = input("Enter operator (+, -, *, /): ")
num2 = int(input("Enter Second Number: "))

if op == '+':
    print("Your sum is:", num1 + num2)
elif op == '-':
    print("Your subtraction is:", num1 - num2)
elif op == '*':
    print("Your multiplication is:", num1 * num2)
elif op == '/':
    # Smart addition: prevent crashing if user divides by zero
    if num2 == 0:
        print("Error: Cannot divide by zero!")
    else:
        print("Your division is:", num1 / num2)
else:
    print("Invalid operator! Please use +, -, *, or /.")
```
