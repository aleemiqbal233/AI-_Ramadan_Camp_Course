```python
# Simple Calculator
num1 = float(input("Enter first number: "))
operator = input("Enter operator (+, -, *, /, **): ")
num2 = float(input("Enter second number: "))

if operator == '+':
    print(f"Result: {num1 + num2}")
elif operator == '-':
    print(f"Result: {num1 - num2}")
elif operator == '*':
    print(f"Result: {num1 * num2}")
elif operator == '/':
    print(f"Result: {num1 / num2}")
elif operator == '**':
    print(f"Result: {num1 ** num2}")
else:
    print("Invalid operator!")
```

```python
my_list = ["apple", "banana", 42, "cherry"]
found = False

for item in my_list:
    if isinstance(item, (int, float)):
        print(f"Numeric value found: {item}")
        found = True
        break 

if not found:
    print("No numeric value found.")
```

```python
customer = {"name": "Ali", "age": 22}
print("Original:", customer)

# Adding a new key 'city'
customer["city"] = "Karachi"
print("Updated:", customer)
```

```python
data = {"val1": 10, "val2": "text", "val3": 25.5, "val4": 100}
total_sum = 0

for value in data.values():
    if isinstance(value, (int, float)):
        total_sum += value

print(f"Sum of numeric items: {total_sum}")
```

```python
arr = [1, 2, 3, 2, 4, 5, 1, 6, 4]
duplicates = []
unique_items = []

for x in arr:
    if x not in unique_items:
        unique_items.append(x)
    elif x not in duplicates:
        duplicates.append(x)

print("Duplicate values are:", duplicates)
```
